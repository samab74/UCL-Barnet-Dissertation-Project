```python
HMO_df = pd.read_csv('HMO Register - 22-01-2024.csv')
```

```python
HMO_df.head()
```

We only want to know their addresses and HMO License

```python
# Keep only the first two columns
HMO_df_subset = HMO_df.iloc[:, :2]

# Display the first few rows of the subset dataset
HMO_df_subset.head()
```

Now just keeping the Postcode

```python
import re

# Define a function to extract the postcode
def extract_postcode(address):
    # Regular expression pattern for UK postcodes
    postcode_pattern = re.compile(r'\b[A-Z]{1,2}[0-9][A-Z0-9]?\s?[0-9][A-Z]{2}\b', re.IGNORECASE)
    match = postcode_pattern.search(address)
    return match.group(0) if match else None

# Apply the function to the HMO ADDRESS column
HMO_df_subset['POSTCODE'] = HMO_df_subset['HMO ADDRESS'].apply(extract_postcode)

# Drop the original HMO ADDRESS column
HMO_df_subset = HMO_df_subset.drop(columns=['HMO ADDRESS'])

# Display the first few rows of the updated dataset
HMO_df_subset.head()
```

```python
HMO_df_subset.to_csv('Final_HMO_df.csv')
```

# Merging the Datasets

```python
# Function to load a CSV file and drop unnamed columns
def load_and_clean_csv(file_path):
    df = pd.read_csv(file_path)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    return df

# Load and clean all the datasets
deprivation_data = load_and_clean_csv('Final_LSOA_deprivation.csv')
economic_activity_data = load_and_clean_csv('Final_LSOA_Economic_Activity.csv')
education_data = load_and_clean_csv('Final_LSOA_Education.csv')
ethnicity_data = load_and_clean_csv('Final_LSOA_Ethnicity.csv')
gender_data = load_and_clean_csv('Final_LSOA_Gender.csv')
population_data = load_and_clean_csv('Final_LSOA_Population.csv')

# Merge all datasets on 'Lower layer Super Output Areas Code'
merged_data = deprivation_data.merge(economic_activity_data, on='Lower layer Super Output Areas Code', suffixes=('_deprivation', '_economic'))
merged_data = merged_data.merge(education_data, on='Lower layer Super Output Areas Code', suffixes=('', '_education'))
merged_data = merged_data.merge(ethnicity_data, on='Lower layer Super Output Areas Code', suffixes=('', '_ethnicity'))
merged_data = merged_data.merge(gender_data, on='Lower layer Super Output Areas Code', suffixes=('', '_gender'))
merged_data = merged_data.merge(population_data, on='Lower layer Super Output Areas Code', suffixes=('', '_population'))

# Ensure only one column for 'Lower layer Super Output Areas'
area_name_columns = [col for col in merged_data.columns if 'Lower layer Super Output Areas' in col and col != 'Lower layer Super Output Areas Code']
merged_data['Lower layer Super Output Areas'] = merged_data[area_name_columns].bfill(axis=1).iloc[:, 0]
merged_data = merged_data.drop(columns=area_name_columns[1:])

# Display the first few rows of the merged dataset to confirm the merge
merged_data.head()
```

```python
# Save the merged dataset to a CSV file
merged_data.to_csv('merged_census_lsoa_data.csv', index=False)
```

# Assigning the polygons to the correct LSOA for Socio-economic Data

```python
import fiona
from shapely.geometry import shape
import pandas as pd

# Lists to hold the properties and geometry
properties = []
geometries = []

with fiona.open('map_converted.shp', 'r') as shapefile:
    for feature in shapefile:
        properties.append(feature['properties'])
        geometries.append(shape(feature['geometry']))

# Convert to DataFrame
shapefile_df = pd.DataFrame(properties)

# Convert geometry to WKT format for easier handling in DataFrame
shapefile_df['geometry'] = [geom.wkt for geom in geometries]

# Display the first few rows of the shapefile DataFrame
shapefile_df.head()
```

```python
census_data = pd.read_csv('merged_census_lsoa_data.csv')

# Merge the shapefile DataFrame with the census data using 'LSOA21CD' as the key
merged_data = pd.merge(shapefile_df, census_data, left_on='LSOA21CD', right_on='Lower layer Super Output Areas Code')

# Display the first few rows of the merged data
merged_data.head()
```

```python
merged_data.to_csv('final_lsoa_census_data.csv', index=False)
```
